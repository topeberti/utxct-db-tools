# Get all samples created with HEXECL material

This notebook joins `samples -> panels -> materials` and filters by material name containing `HEXECL` (case-insensitive).

```python
import dbtools.dbtools as qrs
import pandas as pd
```

```python
# Material filter (edit this if needed)
material_query = 'HEXCEL'

# Load tables with metadata
samples = qrs.get_data_metadata('samples')
panels = qrs.get_data_metadata('panels')
materials = qrs.get_data_metadata('materials')

# Join samples -> panels -> materials
joined = samples.merge(
    panels,
    left_on='panel_id_sample',
    right_on='id_panel',
    how='inner'
)

joined = joined.merge(
    materials,
    left_on='material_id_panel',
    right_on='id_material',
    how='inner'
)

# Filter by material name
samples_hexecl = joined[joined['name_material'].str.contains(material_query, case=False, na=False)].copy()

# Select useful columns
result = samples_hexecl[[
    'id_sample',
    'name_sample',
    'description_sample',
    'id_panel',
    'name_panel',
    'id_material',
    'name_material'
]].sort_values('id_sample')

print(f'Found {len(result)} sample(s) for material query: {material_query}')
result
```
